```python
# import numpy for random selection of files
import numpy as np

# import tensorflow - ensure the version is 2.10.0 or above
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.models import Sequential
print('tensorflow version:', tf.version.VERSION)
from keras.preprocessing.image import ImageDataGenerator

# import certain modules to help with copying of files
import pathlib
import shutil

# import inspect to help creating a dictionary of pre-trained models
import inspect
```

```python
# https://studymachinelearning.com/keras-imagedatagenerator-with-flow_from_directory/
# https://machinelearningmastery.com/how-to-load-large-datasets-from-directories-for-deep-learning-with-keras/

# define the paths to each of train/val/test folders
src_path_train = "../full_data_split/train/"
src_path_val = '../full_data_split/val/'
src_path_test = "../full_data_split/test/"

# instantiate generators for each subset, which can also be used for image augmentation.
train_datagen = ImageDataGenerator()
val_datagen = ImageDataGenerator()
test_datagen = ImageDataGenerator()

#img_height = 180
#img_width = 180
```

```python
batch_size = 8
train_generator = train_datagen.flow_from_directory(
    directory=src_path_train, batch_size=batch_size,
    class_mode='sparse', shuffle=True, seed=42 )

valid_generator = val_datagen.flow_from_directory(
    directory=src_path_val, batch_size=batch_size,
    class_mode='sparse', shuffle=True, seed=42)

test_generator = test_datagen.flow_from_directory(
    directory=src_path_test, batch_size=1,
    class_mode=None, shuffle=False, seed=42)
```

```python
from itertools import islice
def nth_key(dct, n):
    it = iter(dct)
    # Consume n elements.
    next(islice(it, n, n), None) 
    # Return the value at the current position.
    # This raises StopIteration if n is beyond the limits.
    # Use next(it, None) to suppress that exception.
    return next(it)
```
